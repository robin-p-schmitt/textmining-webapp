```python
import numpy as np
import nltk
from nltk.corpus import brown, wordnet as wn
import gensim
import re
import string
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
from mpl_toolkits import mplot3d
import sklearn
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans, DBSCAN
```

```python
# nltk.download('brown')
# brown.categories()
```

```python
def filter(token):
    result = not re.search("[{}\d]".format(string.punctuation), token)
    result &= len(wn.synsets(token)) != 0
    return result
```

```python
sentences = brown.sents(categories = ["adventure"])
sentences = [[token.lower() for token in sent if filter(token)] for sent in sentences]
```

```python
words = brown.words(categories = ["adventure"])
sentences = [[token.lower() for token in words if filter(token)]]
```

```python
word2vec = gensim.models.Word2Vec(sentences, size = 3, iter = 1000, min_count = 5, window = 10)
word2vec.save("word2vec.model")
```

```python
index2word = np.array(word2vec.wv.index2word)
```

```python
len(index2word)
```

```python
kmeans = KMeans(n_clusters = 10).fit(word2vec.wv.vectors)
clusters = [np.array([i for i, cluster in enumerate(kmeans.labels_) if cluster == j]) for j in range(min(kmeans.labels_), max(kmeans.labels_) + 1)]
```

```python
dbscan = DBSCAN(eps = 0.4).fit(word2vec.wv.vectors)
clusters = [np.array([i for i, cluster in enumerate(dbscan.labels_) if cluster == j]) for j in range(min(dbscan.labels_), max(dbscan.labels_) + 1)]
```

```python
len(clusters)
```

```python
num_per_cluster = 100
num_labeled_per_cluster = 10

plt.figure(figsize = (20, 20))
ax = plt.axes(projection='3d')

colors = np.random.choice(np.array(list(mcolors.TABLEAU_COLORS.items()))[:, 0], 10, replace = False)

# go through all clusters
for i, cluster in enumerate(clusters[:10]):
    # select color
    color = colors[i]
    # choose random words to display
    random_instances = np.random.choice(cluster, min(len(cluster), num_per_cluster), replace = False)
    # choose random words to label in the plot
    random_label_instances = np.random.choice(random_instances, min(len(random_instances), num_labeled_per_cluster), replace = False)
    
    # plot some embeddings with a label
    x, y, z = zip(*word2vec.wv.vectors[random_label_instances])
    ax.plot3D(x, y, z, "o", color = color, label = index2word[random_label_instances])
    # plot the remaining embeddings without a label
    x, y, z = zip(*word2vec.wv.vectors[random_instances])
    ax.plot3D(x, y, z, "o", color = color)

plt.legend()
plt.show()
```
